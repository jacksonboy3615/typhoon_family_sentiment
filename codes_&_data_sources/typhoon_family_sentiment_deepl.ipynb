```python
import pandas as pd
import numpy as np
from vaderSentiment.vaderSentiment import SentimentIntensityAnalyzer
import deepl as dp
import os
import re
```

```python
typhoon_family_gpt_df = pd.read_csv('typoon_family_sentiment_gpt.csv')
```

```python
typhoon_family_gpt_df['Reviews'] = typhoon_family_gpt_df['Reviews'].str.replace('\n', '')

typhoon_family_gpt_df['Reviews']
```

```python
auth_key = "your_key"
deepl_client = dp.DeepLClient(auth_key)

# result = deepl_client.translate_text("Hello, world!", target_lang="FR")
```

```python
# Get rid of unnecessary characters

# def clean_text(text):
#     if not isinstance(text, str):
#         return text
#     text = re.sub(r"[^가-힣a-zA-Z0-9\s.,!?]", " ", text)
#     text = re.sub(r"\s+", " ", text).strip()  
#     return text

# typhoon_family_gpt_df["Reviews_Clean"] = typhoon_family_gpt_df["Reviews"].apply(clean_text)

# typhoon_family_gpt_df["Reviews_Clean"].unique()

typhoon_family_gpt_df['Reviews_EN_Deepl'] = typhoon_family_gpt_df['Reviews'].apply(lambda x: deepl_client.translate_text(x, source_lang="KO", target_lang="EN-US").text if type(x) == str else x)
```

```python
typhoon_family_gpt_df['Reviews_EN_Deepl'].unique()
```

```python
typhoon_family_gpt_df
```

```python
sentiment_vader = SentimentIntensityAnalyzer()
typhoon_family_gpt_df['Review_Sentiment_Score_deepl'] = typhoon_family_gpt_df['Reviews_EN_Deepl'].apply(lambda text: sentiment_vader.polarity_scores(text)['compound'])

def sentiment_category_vader(sentiment):
        if sentiment >= 0.05:
            return "Positive"
        elif sentiment <= -0.05:
            return "Negative"
        else:
            return "Neutral"
        
typhoon_family_gpt_df['Review_sentiment_Vader'] = typhoon_family_gpt_df['Review_Sentiment_Score_deepl'].apply(sentiment_category_vader)
```
